```python
#fillna()로 결손 데이터 대체하기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
# 실제 데이터 세트 값이 변경되기 위해서는 다음과 같이 처리해주어야 함 주의
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
#혹은 titanic_df['Cabin'].fillna('C000',inplace=True) 처럼 할 수도 있음
titanic_df.isna().sum()
```

```python
# pandas dataframe 의 lambda식
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
# 나이가 15세 미만이면 'Child', 그렇지 않으면 'Adult'로 구분하는 새로운 칼럼 'Child_Adult' 만들기
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
# 나이가 15세 이하이면 Child, 15~60세 사이는 Adult, 61세 이상은 Elderly로 분류하는 'Age_Cat' 칼럼
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15  else ('Elderly' if 60>=x else 'Adult'))
titanic_df['Age_Cat'].value_counts()
```

```python
# 5세 이하는 Baby, 12살 이하는 Child, 18살 이하는 Teenage, 25살 이하는 Student, 35살 이하는 Young Adult, 60세 이하는 Adult, 그리고 그 이상은 Elderly
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age) : 
    cat = ''
    if age <= 5 : cat = 'Baby'
    elif age <= 12 : cat = 'Child'
    elif age <= 18 : cat = 'Teenage'
    elif age <= 25 : cat = 'Student'
    elif age <= 35 : cat = 'Young Adult'
    elif age <= 60 : cat = 'Adult'
    else : cat = 'Elderly'
    return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_Cat'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_Cat']].head()
```
